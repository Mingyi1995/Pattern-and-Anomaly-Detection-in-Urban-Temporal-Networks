```python
# cols = ['date', 'period', 'start_station', 'end_station', 'population']
# cols = ['日期','時段']
df = pd.read_csv('data/臺北捷運每日分時各站OD流量統計資料_%s.csv'%a,skiprows=[1],sep=' ')
# df09 = pd.read_cvs()
# df = df.rename(columns={'日期         ':'date','時段     ':'period','進站':'start station','出站         ':'end station',
#                        '人次':'amount'})
```

```python
df.dropna(how='all',axis='columns',inplace=True)
```

```python
df.head()
```

```python
name = pd.read_csv('TPE_masternodes_chinses.csv')
```

```python
df_head = df[['日期', 'Unnamed: 1']]
```

```python
df_start = df[['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4','Unnamed: 5', 'Unnamed: 6']]
df_start = df_start.fillna('')
df_start['start'] = df_start['Unnamed: 2'] + df_start['Unnamed: 3'] + df_start['Unnamed: 4'] + df_start['Unnamed: 5'] + df_start['Unnamed: 6']
df_start = df_start[['start']]
```

```python
df_end = df[['Unnamed: 29', 'Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35']]
df_end = df_end.fillna('')
df_end['end'] = df_end['Unnamed: 29'] + df_end['Unnamed: 31'] + df_end['Unnamed: 32'] + df_end['Unnamed: 33'] + df_end['Unnamed: 34'] + df_end['Unnamed: 35']
df_end = df_end[['end']]
```

```python
df_amount = df[['Unnamed: 52',
       'Unnamed: 54', 'Unnamed: 55', 'Unnamed: 56', 'Unnamed: 57',
       'Unnamed: 58', 'Unnamed: 59', 'Unnamed: 60', 'Unnamed: 61',
       'Unnamed: 62', 'Unnamed: 63', 'Unnamed: 64']]
df_amount = df_amount.fillna(0)
df_amount['amount'] = df_amount.apply(lambda x: x.sum(), axis=1)
df_amount = df_amount[['amount']]
```

```python
df_1 = df_head.merge(df_start, left_index=True, right_index=True)
df_2 = df_end.merge(df_amount, left_index=True, right_index=True)
df_compact = df_1.merge(df_2, left_index=True, right_index=True)
df_compact.head()
```

```python
df_compact = df_compact.rename(columns={'日期':'data','Unnamed: 1':'period'})
```

```python
df_compact.to_csv('compact/%s.csv'%a,sep=',')
```

```python
chinese_name = list(name['Chinese_name'])
stop_id = list(name['stop_id'])
new_converter = {}
for i in range(len(chinese_name)):
    new_converter[chinese_name[i]] = stop_id[i]
```

```python
name_convert = lambda x: new_converter[x] 
df = df[df['amount']!=0]
df['start_id'] = df['start'].apply(name_convert)
df['end_id'] = df['end'].apply(name_convert)
```
