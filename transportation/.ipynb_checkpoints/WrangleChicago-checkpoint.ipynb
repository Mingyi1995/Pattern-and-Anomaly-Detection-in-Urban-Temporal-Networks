```python
import pandas as pd
```

```python
df = pd.read_csv('DCEdgeDatewiseAggregated.csv')
```

```python
df = df[['date','start_id','end_id','amount']]
```

```python
df['filter'] = df['date'].apply(lambda x:str(x.split('/')[-1]))
df = df[(df['filter']=='2017')|(df['filter']=='2018')]
del df['filter']
```

```python
len(df)
```

```python
df.to_csv('DCEdgeDatewiseAggregated.csv',index=False)
```

```python
Grouped = df.groupby(['start_id','end_id'], as_index=False).sum()
```

```python
Grouped.to_csv('DCEdgeYearwiseAggregated.csv',index=False)
```
